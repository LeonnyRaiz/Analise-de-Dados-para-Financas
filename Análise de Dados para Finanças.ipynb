```python
#Libs necessárias

#Libs para modelagem e matrizes

import numpy as np
import pandas as pd

#Libs para análises gráficas

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

#Lib para ignorar avisos

import warnings

#Desabilitando avisos

warnings.filterwarnings('ignore')
```

```python
#Lendo os dados

Base_Dados = pd.read_excel('Vase_004 - Magalu - Sem Resolução.xlsx')

#Verificando

Base_Dados.head()
```

```python
Base_Dados.shape
```

```python
Base_Dados.info()
```

```python
#Series Temporais

#Definindo a data como index

Dados = Base_Dados.set_index('Data')

Dados.head()
```

```python
#Gerando um gráfico com o valor do fechamento das ações durante o ano
plt.style.use('seaborn-darkgrid')
plt.figure(figsize =(16, 5))
plt.title('Análise das ações da Magalu - Fechamento', fontsize = 14, loc = 'left')
plt.plot(Dados.index, Dados['Fechamento'])
plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da Ação(R$)');
```

```python
#Últimos registros da base de dados
Dados.tail()
```

```python
Media_Movel = Dados['Fechamento'].rolling(5).mean()
Media_Tendencia = Dados['Fechamento'].rolling(30).mean()

plt.style.use('seaborn-darkgrid')
plt.figure(figsize =(16, 5))
plt.title('Análise das ações da Magalu - Fechamento', fontsize = 14, loc = 'left')

plt.plot(Dados.index, Dados['Fechamento'])
plt.plot(Dados.index, Media_Movel)
plt.plot(Dados.index, Media_Tendencia)

plt.xlabel('Periodo da cotação')
plt.ylabel('Valor da Ação(R$)');
```

```python
#Boxplot Mensal

Base_Dados['Mes'] = Base_Dados['Data'].dt.month

plt.figure(figsize=(16, 5))
sns.boxplot(data = Base_Dados, x= 'Mes', y = 'Fechamento')
```

```python
Base_Dados.groupby(['Mes']).describe()['Fechamento']
```

```python
# Gráfico interativo para saber se a ação fechou o dia em alta ou em baixa

Grafico = go.Figure(
    data=[
          go.Candlestick(
              x= Dados.index,
              open = Dados['Abertura'],
              high = Dados['Maior'],
              low = Dados['Menor'],
              close = Dados['Fechamento'],
          )
    ]
)

Grafico.update_layout( xaxis_rangeslider_visible=False )

Grafico.show()
```

# Insights

**1-** Segundo os gráficos as ações da Magalu tiveram uma queda quase que constante ao longo dos meses, chegando a fechar o ano valendo por volta de 6 reais, portanto quem comprou no começo do ano teve um prejuízo altíssimo

**2-** De acordo com o último gráfico que simula o gráfico da Ibovespa, as ações tiveram seu fechamento na grande maioria das vezes em negativo, portanto é um investimento de alto risco comprar as ações no momento
